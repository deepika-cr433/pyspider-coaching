```python
df1['Directors']=df1['Directors'].fillna("unknown")
```

```python
df1['Directors'].isnull()
```

```python
df1['Rotten Tomatoes']=df1['Rotten Tomatoes'].ffill()
```

```python
df1['Rotten Tomatoes'] = df1['Rotten Tomatoes'].str.rstrip('%').astype(int)
```

```python
df1['Rotten Tomatoes'] = df1['Rotten Tomatoes']/10
```

```python
df1.drop(columns='Type',inplace=True)
```

```python
df1['IMDb']=pd.to_numeric(df1['IMDb']).fillna(0).astype(int)
```

```python
df1['Language']=df1['Language'].fillna('Unknown')
df1['Country']=df1['Country'].fillna("Unknown")
df1['Genres']=df1['Genres'].fillna('Unknown')
df1['Age']=df1['Age'].fillna(0)
```

```python
df1['Runtime']=df1['Runtime'].fillna(0)
df1['Runtime']=df1['Runtime'].astype(int)
```

```python
pd.unique(df1['Runtime'])
```

```python
df1.info()
```

```python
df1["Runtime"]=pd.cut(df1['Runtime'],bins=[0,60,180,400],labels=['short','long','extra_long'],right=False)
df1.rename(columns={'Runtime':'Runtime_category'},inplace=True)
df1
```

```python
pd.unique(df1['Runtime_category'])
df1['Runtime_category']=df1['Runtime_category'].fillna('Unknown')
```
