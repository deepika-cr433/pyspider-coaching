```python
df1['Age']=df1['Age'].ffill()
df1['Directors']=df1['Directors'].bfill()
```

```python
df1['IMDb']=df1['IMDb'].fillna(5.902751)
df1['Runtime']=df1['Runtime'].bfill()
```

```python
df1['Genres'].mode()
df1['Country'].mode()
df1['Language'].mode()
```

```python
df1['Genres']=df1['Genres'].fillna('Drama')
df1['Country']=df1['Country'].fillna('United States')
df1['Language']=df1['Language'].fillna('English')

```

```python
df1=df1.dropna(axis=1)
```
