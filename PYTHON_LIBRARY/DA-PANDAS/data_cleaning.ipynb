FILLER METHODS
fillna--------------> ffill(),bfill()
dropna

```python
df['age'].fillna(29.881135,inplace=True)
```

```python
df['fare']=df['fare'].fillna(method='ffill')
```

```python
df['embarked'].mode()
```

```python
df['embarked']=df['embarked'].fillna('S')
```

```python
df=df.dropna(axis=1)
```

```python
df['survived']=df['survived'].astype('Float64')
```

**feature engineering**

```python
df["embarked"]=df["embarked"].replace({"S":"Southampton","C":"Cherbourg","Q":"Queenstown"})
df
```

```python
df['family_count']=df['sibsp']+df['parch']
df
```

```python
df['name'].str.split(',',expand=True)
df['last_name']=df['name'].str.split(',',n=1,expand=True)[0]
df['title_fname']=df['name'].str.split(',',n=1,expand=True)[1]
df
```

```python
x,y=df['title_fname'].str.split('.',n=1,expand=True)
print(x)
print(y)
```

```python
df['title']=df['title_fname'].str.split('.',n=1,expand=True)[0]
df['first_name']=df['title_fname'].str.split('.',n=1,expand=True)[1]
df
```
