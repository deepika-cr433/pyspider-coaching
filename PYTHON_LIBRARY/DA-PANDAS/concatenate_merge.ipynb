# concatenate

```python
import pandas as pd
d1={
    'empno':['e1','e2','e3','e4','e5'],
    'name':['smith','scott','allen','smith','Martin'],
    'sal':[1000,2000,1000,5000,2000],
    'hiredate':['01-jan-2000','01-jan-2001','4-feb-2000','8-dec-2001','01-jan-2001'],
    'deptno':[10,10,30,10,20]
}
d2={
    'deptno':[10,20,30,40],
    'dname':['accounting','research','sales','operations'],
    'loc':['New York','Dallas','Chicago','Dallas']
}
emp=pd.DataFrame(d1)
dept=pd.DataFrame(d2)
print(emp)
print(dept)
```

```python
pd.concat([emp,dept],keys=['employee','department'])#default axis=0
```

```python
pd.concat([emp,dept],axis=1,keys=['employees','department'])
# here deptno in department table is in float datatype bcz it as NaN value.
```

```python
pd.concat([emp,dept],axis=1)
```

```python
#-------------------------------------------------optional
emp.set_index('empno',inplace=True)
dept.set_index('deptno',inplace=True)
```

```python
print(emp)
print(dept)
```

```python
pd.concat([emp,dept],ignore_index=True)# ignore_index=true means from both table it ingnore the index
```

# merge()

```python
pd.merge(emp,dept,on='deptno')
```

```python
pd.merge(emp,dept,how='cross')# for cross join on shouldn't be used.. means no joining condition
#we can use 'outer','left,'right','inner'.
```

```python
emp.rename(columns={'deptno':'dno'},inplace=True)
df3=pd.merge(emp,dept,left_on='dno',right_on='deptno')
#emp--dno and dept--deptno
df3
```

```python
pd.merge(emp,dept,how='left')
```
