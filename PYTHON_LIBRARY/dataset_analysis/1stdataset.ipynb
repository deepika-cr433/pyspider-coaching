Which countries produce the most movies on Netflix?

```python
plt.figure(figsize=(10,6))
movie_df = df[df['type'] == 'Movie'] 
country_counts = movie_df['country'].value_counts() 
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10])
```

Which countries produce the most TV shows on Netflix?

```python
plt.figure(figsize=(10,6))
tv_show_df = df[df['type'] == 'TV Show'] 
country_counts = tv_show_df['country'].value_counts() 
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10])
```

What is the relationship between release year and content rating?

Which genres are most popular among movies?

```python
plt.figure(figsize=(20,6))
from collections import Counter
movie_df = df[df['type'] == 'Movie'] 
all_genres = list(movie_df['listed_in'].str.split(',').explode()) 
top_genres = Counter(all_genres).most_common(10) 
sns.barplot(x=[genre[0] for genre in top_genres], y=[genre[1] for genre in top_genres])
```

Which genres are most popular among TV shows?

```python
plt.figure(figsize=(20,6))
tv_show_df = df[df['type'] == 'TV Show']
all_genres = list(tv_show_df['listed_in'].str.split(',').explode())
top_genres = Counter(all_genres).most_common(10)
sns.barplot(x=[genre[0] for genre in top_genres], y=[genre[1] for genre in top_genres])
```

How does the number of cast members vary between movies and TV shows?
